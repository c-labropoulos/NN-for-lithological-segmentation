```python
    !pip  install pyshp
     #Plot map from shapefile  using using  different colors for different shapes
    import shapefile
    import matplotlib.pyplot as plt
    
    #Read the shapefile
    shp_path = '/content/drive/MyDrive/Lithology_Dataset/Petroto/Petroto.shp'
    sf = shapefile.Reader(shp_path)
    
    #Get the shapes and records
    shapes = sf.shapes()
    records = sf.records()
    
    #Get the shapefile bounding box
    minx = sf.bbox[0]
    miny = sf.bbox[1]
    maxx = sf.bbox[2]
    maxy = sf.bbox[3]
    
    #Create a figure
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    #Set the axis limits
    ax.set_xlim(minx, maxx)
    ax.set_ylim(miny, maxy)
    
    #Get the unique values of the field
    unique_values = set([record[0] for record in records])
    
    #Create a dictionary to map values to colors
    colormap = dict(zip(unique_values, ['red', 'green', 'blue']))
    
    #Plot the shapes
    for shape in shapes:
        x = [i[0] for i in shape.points[:]]
        y = [i[1] for i in shape.points[:]]
        plt.plot(x, y, cmap="RdYlBu")
    
    #Show the plot
    plt.show()
    
```
