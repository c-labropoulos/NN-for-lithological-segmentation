```python
import os
from osgeo import ogr
!pip install geopandas
import geopandas as gpd


def shp_2_json(shapefiles_path, output_path) :
    """
    Function transforms given directory with shapefiles into geojsons.
    
    :param shapefiles_path: (str) Path to the directory with shapefiles.
    :param output_path: (str) Path to the output geojson file.
    """
    
    shp = '.shp'
    geojson = 'json'
    shapefiles = [fname for fname in os.listdir(shapefiles_path) if fname.endswith(shp)]
    
    for sfile in shapefiles:
        # Get and create paths
        spath = os.path.join(shapefiles_path, sfile)
        gjfile = sfile[:-3]+geojson
        gjpath = os.path.join(output_path, gjfile)
        
        # Load & save
        gdf = gpd.read_file(spath)
        gdf.to_file(gjpath, driver='GeoJSON')
#shppath='/content/drive/MyDrive/LithDatasSHP'
#out='/content/drive/MyDrive/GeojsonLithData'
#shp_2_json(shppath,out)
```

```python
import gdal
import ogr
import os
import numpy as np

#Defining the mask function
def mask(raster_path, mask_path, out_path):
    #Open the raster and mask files
    raster = gdal.Open(raster_path)
    mask = ogr.Open(mask_path)

    #Get the raster and mask layer
    raster_layer = raster.GetRasterBand(1)
    mask_layer = mask.GetLayer()

    #Get the raster geotransform
    raster_gt = raster.GetGeoTransform()

    #Create an array the size of the raster
    array = np.zeros((raster_layer.YSize, raster_layer.XSize))

    #Loop through the mask features
    for feature in mask_layer:
        #Get the geometry of the feature
        geom = feature.GetGeometryRef()
        #Get the bounding box of the feature
        min_x, max_x, min_y, max_y = geom.GetEnvelope()
        #Convert the bounding box to pixel coordinates
        px_min = int((min_x - raster_gt[0]) / raster_gt[1])
        px_max = int((max_x - raster_gt[0]) / raster_gt[1])
        py_min = int((min_y - raster_gt[3]) / raster_gt[5])
        py_max = int((max_y - raster_gt[3]) / raster_gt[5])
        #Loop through the pixels in the bounding box
        for py in range(py_min, py_max):
            for px in range(px_min, px_max):
                px_dist = px * raster_gt[1] + raster_gt[0]
                py_dist = py * raster_gt[5] + raster_gt[3]
                #Get the pixel's coordinate
                p = ogr.Geometry(ogr.wkbPoint)
                p.AddPoint(px_dist, py_dist)
                #If the point is within the polygon, set the pixel to 1
                if geom.Contains(p):
                    array[py, px] = 255

    #Create the output raster
    driver = gdal.GetDriverByName('GTiff')
    out_raster = driver.Create(out_path, raster_layer.XSize, raster_layer.YSize, 5, gdal.GDT_Byte)
    out_raster.SetGeoTransform(raster_gt)
mask('/content/drive/MyDrive/LithDatasTIF/Amfilochia.tif', '/content/drive/MyDrive/LithDatasSHP/Amfilochia.shp', '/content/drive/MyDrive/LithDatasTIF/MaskedAmfilochia.tif')
```

```python
"""
image segmentaion for satellite images
"""
import os
import glob
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from keras.models import Model
from keras.layers import Input, Conv2D, MaxPooling2D, Dropout, UpSampling2D
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint, LearningRateScheduler
from keras import backend as K
from skimage.io import imread
from sklearn.metrics import jaccard_score
from keras.preprocessing.image import ImageDataGenerator

from keras.layers import concatenate
from keras.callbacks import EarlyStopping

# Set some parameters
im_width = 128
im_height = 128
border = 5
path_train = './data/train/'
path_test = './data/test/'

# Get and resize train images and masks
def get_data(path, train=True):
    ids = next(os.walk(path + "images"))[2]
    X = np.zeros((len(ids), im_height, im_width, 3), dtype=np.uint8)
    sizes = []
    print('Getting and resizing images ... ')
    for n, id_ in enumerate(ids):
        img = imread(path + '/images/' + id_)[:,:,:3]
        sizes.append([img.shape[0], img.shape[1]])
        img = cv2.resize(img, (im_width, im_height))
        X[n] = img
    print('Done!')
    if train:
        y = np.zeros((len(ids), im_height, im_width, 1), dtype=np.bool)
        print('Getting and resizing masks ... ')
        for n, id_ in enumerate(ids):
            mask = imread(path + '/masks/' + id_)
            mask = cv2.resize(mask, (im_width, im_height))
            mask = np.expand_dims(mask, axis=-1)
            y[n] = mask
        print('Done!')
        return X, y, sizes
    else:
        return X, sizes
```
